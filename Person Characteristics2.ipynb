```python
# Import the necessary Libraries for the Hierarchical Clustering
from pandas import read_excel
import numpy as np
import matplotlib.pyplot as plt
import scipy.cluster.hierarchy as shc
from sklearn.cluster import AgglomerativeClustering
```

```python
# Load the Dataset
df = read_excel('features.xlsx', sheet_name = 2)
df.shape
```

```python
df.head()
```

```python
data = df.iloc[:, 1:7].values
data.shape
```

```python
plt.figure(figsize=(10, 7))
plt.title("Persons Dendograms")
plt.xlabel("Cluster Size")
plt.ylabel("Distance")
dend = shc.dendrogram(shc.linkage(data, method='ward'))
```

```python
plt.figure(figsize=(10, 7))
plt.title("Persons Dendograms")
plt.xlabel("Cluster Size")
plt.ylabel("Distance")
dend = shc.dendrogram(shc.linkage(data, method='ward'))
plt.axhline(y=5, color='r', linestyle='--')
```

```python
cluster = AgglomerativeClustering(n_clusters=2, affinity='euclidean', linkage='ward')
cluster.fit_predict(data)
```

```python
cluster = AgglomerativeClustering(n_clusters=2, affinity='euclidean', linkage='complete')
cluster.fit_predict(data)
```

```python
cluster = AgglomerativeClustering(n_clusters=2, affinity='euclidean', linkage='average')
cluster.fit_predict(data)
```

```python
cluster = AgglomerativeClustering(n_clusters=2, affinity='manhattan', linkage='complete')
cluster.fit_predict(data)
```

```python
N = 83
plt.figure(figsize=(17,10))
plt.grid(True)
plt.scatter(data[:,0], data[:,3], s=(30 * np.random.rand(N))**2, c=cluster.labels_, cmap='rainbow',alpha=0.5)
plt.scatter(data[:,1], data[:,4], s=(30 * np.random.rand(N))**2, c=cluster.labels_, cmap='rainbow',alpha=0.5)
plt.scatter(data[:,2], data[:,5], s=(30 * np.random.rand(N))**2,  c=cluster.labels_, cmap='rainbow',alpha=0.5)
plt.show()
```
