```python
# Import the necessary Libraries for the Hierarchical Clustering
from pandas import read_excel
import numpy as np
import matplotlib.pyplot as plt
import scipy.cluster.hierarchy as shc
from sklearn.cluster import AgglomerativeClustering
```

```python
# Load the Dataset
df = read_excel('features.xlsx', sheet_name = 2)
df.shape
```

```python
df.head()
```

```python
# data = df.iloc[:, 1:7].values
c1 = df.iloc[:, [1,4]].values
c2 = df.iloc[:, [2,5]].values
c3 = df.iloc[:, [3,6]].values
data = np.concatenate((c1, c2, c3))
data.shape
```

```python
plt.figure(figsize=(10, 7))
plt.title("Persons Dendograms")
plt.xlabel("Cluster Size")
plt.ylabel("Distance")
dend = shc.dendrogram(shc.linkage(data, method='ward'))
```
